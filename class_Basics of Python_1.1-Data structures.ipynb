## Sets
A set is a collection which is unordered and unindexed. 
In Python sets are written with curly brackets.

```python
set1={1,3,8,9,4,3,5,2,1}
print(set1,"\n")
set2={'A','C','D','A','X'}
print(set2)
```

```python
set2={'A','D','C','X'}
print(set2)
```

```python
# Access items
# You cannot access items in a set by referring to an index, 
# since sets are unordered the items has no index.
#set1[3]
```

```python
a=set()
type(a)
```

```python
# Access items
# loop through the set items using a for loop
for j in set1:
    print(j)
```

```python
print('D' in set2)
```

## Methods
- add(): Adds an element to the set
- clear(): Removes all the elements from the set
- copy(): Returns a copy of the set
- difference(): Returns a set containing the difference between two or more sets
- difference_update(): Removes the items in this set that are also included in another, specified set
- discard(): Remove the specified item
- intersection(): Returns a set, that is the intersection of two other sets
- intersection_update(): Removes the items in this set that are not present in other, specified set(s)
- isdisjoint(): Returns whether two sets have a intersection or not
- issubset(): Returns whether another set contains this set or not
- issuperset(): Returns whether this set contains another set or not
- len(): Finds length of set
- pop(): Removes an element from the set
- remove(): Removes the specified element
- symmetric_difference(): Returns a set with the symmetric differences of two sets
- symmetric_difference_update(): inserts the symmetric differences from this set and another
- union(): Return a set containing the union of sets
- update(): Update the set with the union of this set and others

```python
# add(): Adds an element to the set
print(set1)
set1.add(12)
print(set1)
```

```python
# clear(): Removes all the elements from the set
set1.clear()
print(set1)
```

```python
# copy(): Returns a copy of the set
set1=set2.copy()
print(set1)
```

```python
# difference(): Returns a set containing the difference between two or more sets
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

z = x.difference(y)
print(z)
```

```python
# difference_update(): Removes the items in this set that are also included in another, specified set
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

x.difference_update(y)
print(x)
```

```python
y
```

```python
# discard(): Remove the specified item
print(set1)
set1.discard('X')
print(set1)
```

```python
# intersection(): Returns a set, that is the intersection of two other sets
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

z = x.intersection(y)
print(z)
```

```python
# intersection_update(): Removes the items in this set that are not present in other, specified set(s)
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

x.intersection_update(y)
print(x)
print(y)
```

```python
# isdisjoint(): Returns whether two sets have a intersection or not
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

z = x.isdisjoint(y)
print(z)
```

```python
# issubset(): Returns whether another set contains this set or not
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple",'banana','cherry'}

z = x.issubset(y)
print(z)
```

```python
# issuperset(): Returns whether this set contains another set or not
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple",'banana','cherry'}

z = y.issuperset(x)
print(z)
```

```python
# len(): Finds length of set
len(y)
```

```python
# pop(): Removes an element from the set and returns the removed value.
x = { "banana","apple","cherry"}
print(x)
print(x.pop())
print(x)
```

```python
#x = {"apple", "banana", "cherry"}
#print(x)
```

```python
# remove(): Removes the specified element
y={'samsung', 'apple', 'mi'}
print(y)
y.remove('mi')
#y.remove('MI')
#y.discard('MI')
print(y)
```

```python
# symmetric_difference(): Returns a set with the symmetric differences of two sets
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

z = x.symmetric_difference(y)
print(z)
```

```python
# symmetric_difference_update(): inserts the symmetric differences from this set and another
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

y.symmetric_difference_update(x)
print(y)
```

```python
# union(): Return a set containing the union of sets
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

z=x.union(y)
print(z)
```

```python
# update(): Update the set with the union of this set and others
x = {"apple", "banana", "cherry"}
y = {"samsung", "mi", "apple"}

x.update(y)
print(x)
print(y)
```

# Methods
- clear(): Removes all the elements from the dictionary
- copy(): Returns a copy of the dictionary
- fromkeys(): Returns a dictionary with the specified keys and values
- pop(): Removes the element with the specified key
- popitem(): Removes the last inserted key-value pair
- setdefault(): Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
- update(): Updates the dictionary with the specified key-value pairs

```python
# clear(): Removes all the elements from the dictionary
print(d1)
d1.clear()
print(d1)
```

```python
# pop(): Removes the element with the specified key
print(d1)
d1.pop('Name')
print(d1)
```

```python
# setdefault(): Returns the value of the specified key. 
# If the key does not exist: insert the key, with the specified value
d1={'Name': 'Ankit', 'Class': '12th', 'Roll No.': 25, 'Result': 'Pass'}
print(d1)
d2=d1.setdefault("Class")
print(d2)
d3=d1.setdefault("Location",'Pune')
print(d3)
```
